## Association Rule - Modeling

위치데이터, 설문정보, 앱데이터를 모두 활용한 연관규칙 분석

<br>

```python
# min_support 0.01 이하는 발생하지 않은 사건으로 취급
# confidence는 일단 고려하지 않음
# lift의 min_threshold를 2로 설정 (두 사건이 독립일때 기대되는 동시발생빈도보다 2배 높은 동시발생빈도)

frequent_itemsets = apriori(AR_df, min_support=0.01, use_colnames=True, max_len=2)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
```

```python
# 총 5488개의 규칙 발견, 조건절과 결과절이 서로 뒤바뀐 경우라면 같은 경우로 취급하기 때문에 이보다 규칙 수는 더 적음

len(rules)
```

```python
# 중복 규칙을 삭제하는 과정

dup_rule_df = rules.sort_values(by='lift', ascending=False)[['antecedents','consequents']]
dup_rule_df['antecedents'] = dup_rule_df['antecedents'].map(lambda x: str(x).replace('frozenset({','').replace('})','')[1:-1])
dup_rule_df['consequents'] = dup_rule_df['consequents'].map(lambda x: str(x).replace('frozenset({','').replace('})','')[1:-1])

rule_set = []

for i in range(len(dup_rule_df)):
    x = set(dup_rule_df.loc[i,:])
    rule_set.append(x)

rules['dup?'] = rule_set
rules['dup?'] = rules['dup?'].map(lambda x: str(x))
rules = rules.drop_duplicates(subset=['dup?']).sort_values(by='lift', ascending=False)

del rules['dup?']
```

```python
rules_for_network.columns = ['Source','Target', 'Weight']
rules_for_network.head()
```

```python
# 분석에 사용할 lift 값만 따로 edgelist형태로 저장

rules_for_network.to_csv('AR_allcol_edgelist.csv', index=False)
```

```python
# 혹시 몰라서 매트릭스 형태로도 저장

AR_adj_matrix = pd.DataFrame(index=AR_df.columns, columns=AR_df.columns)

for i,j,k in zip(rules_for_network['antecedents'], rules_for_network['consequents'], rules_for_network['lift']) :
    AR_adj_matrix.loc[i,j] = k
    AR_adj_matrix.loc[j,i] = k   
    
AR_adj_matrix_fillzero = AR_adj_matrix.fillna(int(0))
AR_adj_matrix_fillzero.to_csv("AR_adj_matrix_fillzero.csv")
```

### 3. 앱데이터 라벨링

앱 사용패턴 - <br>
앱 - 카테코리에 따라 라벨링

<br>

```python
AR_App.head()
```

```python
# 대상 앱과 그 카테고리

target_app_cat = pd.read_csv("target_app_cat.csv")
target_app_cat.head()
```

```python
# App_num(사용하는 앱의 수) / Usage_time(사용시간) / app_category(가장 선호하는 앱 카테고리) / 개별 앱 고유 카테고리 로 분류

Appdata_label = pd.DataFrame(data = AR_App.columns).rename(columns={0:'Colname'})[1:]
Appdata_label['Label'] = Appdata_label['Colname'].map(lambda x: 'App_number' if "App_num" in x 
                                     else 'Usage_time'if "Usage_time" in x 
                                     else 'Favorite_category'if "app_category" in x
                                     else None)

Appdata_label = Appdata_label.merge(target_app_cat, how='left', left_on='Colname', right_on='app_name')
Appdata_label_1 = Appdata_label[lambda x: x['Label'].notnull()]
Appdata_label_1 = Appdata_label_1[['Colname','Label']]

Appdata_label_2 = Appdata_label[lambda x: x['Label'].isnull()]
Appdata_label_2 = Appdata_label_2[['Colname','app_category']]
Appdata_label_2.columns = ['Colname','Label']

Appdata_label = pd.concat([Appdata_label_1,Appdata_label_2])
```

```python
Appdata_label.head()
```

```python
# concat 후 저장

allcol_label = pd.concat([locdata_label,PSdata_label,Appdata_label])
allcol_label.columns = ['id','category']
allcol_label.to_csv("allcol_label.csv", index=False)
```
