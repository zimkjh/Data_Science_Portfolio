```python
# pay_data 컬럼 확인

print("pay_data column length :",len(pay_data.columns))

pay_data.columns
```

```python
# Area_Name1, Area_Code1 1대1 대응 표
# Area_Code1에는 null 값 없음

pay_data_area_name_1 = pay_data.groupby(["Area_Name1", "Area_Code1"]).count()\
    .reset_index()[["Area_Name1", "Area_Code1"]].set_index("Area_Code1").T
pay_data_area_name_1
```

```python
# Area_Name 2, Area_Code2 1대1 대응 표
# Area_Code2에는 null 값 없음

pay_data_area_name_2 = pay_data.groupby(["Area_Name2", "Area_Code2"]).count()\
    .reset_index()[["Area_Name2", "Area_Code2"]].set_index("Area_Code2").T
pay_data_area_name_2
```

```python
# Area code, name dictionary 생성

dict_area_1 = pay_data_area_name_1.to_dict()
dict_area_2 = pay_data_area_name_2.to_dict()
```

```python
# PAY_APPROVAL_TYPE column 만들기

pay_data["PAY_APPROVAL_TYPE"] = pay_data["APPROVAL_TYPE"].apply(lambda x : x[0])
```

```python
# PAY_CATEGORY_CODE column 만들기

pay_data["PAY_CATEGORY_CODE"] = pay_data["CATEGORY_CODE"]
```

```python
# PAY_SMS_REGISTRATION_MONTH column 만들기

pay_data["PAY_SMS_REGISTRATION_MONTH"] = pay_data["SMS_REGISTRATION_MONTH"].apply(lambda x: str(x)[-2:])
# 다만 이번 데이터는 모두 11월이다
```

```python
# PAY_SMS_REGISTRATION_DATA column 만들기

pay_data["PAY_SMS_REGISTRATION_DATE"] = pay_data["SMS_REGISTRATION_DATE"].apply(lambda x: str(x)[-2:])
# 다만 이번 데이터는 모두 11월이다
```

```python
# PAY_SMS_REGISTRATION_TIME column 만들기
# 시간만 사용

pay_data["PAY_SMS_REGISTRATION_TIME"] = pay_data["SMS_REGISTRATION_TIME"].apply(lambda x: str(x)[:-4])
```

```python
# PAY_SMS_REGISTRATION_TIME column 만들기

pay_data["PAY_CARD_PAYMENT_TYPE"] = pay_data["CARD_PAYMENT_TYPE"]
```

```python
# PAY_APPROVAL_METHOD column 만들기
# = 할부개월수

pay_data["PAY_APPROVAL_METHOD"] = pay_data["APPROVAL_METHOD"]
```

```python
# PAY_APPROVAL_METHOD column 만들기
# = 할부개월수

pay_data["PAY_PRICE"] = pay_data["PRICE"]
```
