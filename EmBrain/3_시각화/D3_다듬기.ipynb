### csv 파일을 json 형식으로 바꾸기

```python
n = pd.read_csv("nodes.csv").drop("Unnamed: 0", axis =1)
e = pd.read_csv("edges.csv").drop("Unnamed: 0", axis = 1)
```

```python
def toJson(df):
    # 모든 변수 string화
    df = df.astype(str)
    df_col = df.columns
    for i in range(len(df_col)):
        if(i == 0):
            df[df_col[i]] = ['{ '+df_col[i]+': "'+x+'", ' for x in df[df_col[i]]]
        elif(i == len(df_col)-1):
            df[df_col[i]] = [df_col[i]+': "'+x+'" },' for x in df[df_col[i]]]
        else:
            df[df_col[i]] = [df_col[i]+': "'+x+'", ' for x in df[df_col[i]]]
    # index 없애기
    df.set_index(df_col[0],inplace = True)      
    return df.to_string().replace("    ", " ").replace("   ", " ").replace("  ", " ")
```
